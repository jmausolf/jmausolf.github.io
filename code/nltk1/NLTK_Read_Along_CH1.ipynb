```python
def count_percent_key(text, key):
    count = text.count(key)
    percent = count / len(text)
    print ("The word ", key, "is mentioned: ", count, "times.", '\n', "It represents ", percent, "percent of the total words.")

count_percent_key(text2, "love")
    
```

```python
def long_words(text):
    V = set(text)
    lw = [w for w in V if len(w) > 15]
    print (sorted(lw))
    
long_words(text2)
```

```python
long_words(text1)
```

```python
long_words(text4)
```

```python
def impt_words(text, l, k):
    fdist5 = FreqDist(text)
    print (sorted(w for w in set(text) if len(w) > l and fdist5[w] > k))
    
impt_words(text1, 8, 35)
```

```python
impt_words(text2, 8, 35)
```

```python
impt_words(text3, 5, 15)
```
